```python
import pandas as pd
import json

# Initialize an empty list to store JSON objects
data_list = []

# Open the JSON file and read line by line
with open('/home/nikhil/Downloads/MusicBench_train.json', 'r') as json_file:
    for line in json_file:
        try:
            # Parse each JSON object
            data = json.loads(line)
            data_list.append(data)
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON: {e}")

# Convert the list of JSON objects to a DataFrame
df = pd.DataFrame(data_list)

# Save the DataFrame to a CSV file
df.to_csv('/home/nikhil/Downloads/MusicBench_train.csv', index=False)
```

```python
import os
import shutil
import pandas as pd

class DatabaseCreator:
    def __init__(self, wav_folder_path, csv_path):
        self.wav_folder_path = wav_folder_path
        self.csv_path = csv_path
        self.training_data_path = 'training_data'
        os.makedirs(self.training_data_path, exist_ok=True)

    def process_csv(self, limit=40000):
        df = pd.read_csv(self.csv_path)
        df = df.head(limit)
        c = 0
        # Iterate over rows to copy files and create text files
        for index, row in df.iterrows():
            ytid = row['location']
            caption = row['main_caption']
            wav_filename = ytid.split("/")[1]
            wav_file_path = os.path.join(self.wav_folder_path, wav_filename)
            if c ==0:
                print(ytid)
                print(caption)
                print(wav_filename)
                print(wav_file_path)
            c += 1
            if os.path.exists(wav_file_path):
                # Copy WAV file to training_data folder
                shutil.copy(wav_file_path, self.training_data_path)
                # Create text file with caption
                txt_filename = f'{ytid.split(".")[0].split("/")[1]}.txt'
                print(txt_filename)
                txt_file_path = os.path.join(self.training_data_path, txt_filename)
                with open(txt_file_path, 'w') as txt_file:
                    txt_file.write(caption)
            print(index)
        print("Done")


if __name__ == '__main__':
    wav_folder_path = "/home/nikhil/Documents/yeshiva_projects/audiocraft/MusicBench/datashare/data_aug2"
    csv_path = "MusicBench_train.csv"
    creator = DatabaseCreator(wav_folder_path, csv_path)
    creator.process_csv(limit=5000)
```

```python
def find_unique_element(arr):
    unique_element = 0
    for num in arr:
        unique_element ^= num
        print(unique_element)
    return unique_element
```
